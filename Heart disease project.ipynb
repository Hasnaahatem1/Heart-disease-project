# Part3-Preprocessing & Encoding Decisions

### 9- Handling Missing Values 

```python
df['BMI'] = df['BMI'].fillna(
    df.groupby('HeartDisease')['BMI'].transform('median')
)

df['SleepTime'] = df['SleepTime'].fillna(
    df.groupby('HeartDisease')['SleepTime'].transform('median')
)

df['SkinCancer'] = df['SkinCancer'].fillna(
    df.groupby('HeartDisease')['SkinCancer']
      .transform(lambda x: x.mode()[0])
)

df['DiffWalking'] = df['DiffWalking'].fillna(
    df.groupby('HeartDisease')['DiffWalking']
      .transform(lambda x: x.mode()[0])
)
```

### 10-Feature Type Classification

|Feature            | Type                    | Justification |
|--------------------|-------------------------|---------------|
| HeartDisease       | Binary                  | Target variable with two values (Yes / No). |
| BMI                | Continuous numerical    | Numeric values representing body mass index. |
| Smoking            | Binary                  | Two possible values (Yes / No). |
| AlcoholDrinking    | Binary                  | Two possible values (Yes / No). |
| Stroke             | Binary                  | Indicates whether the person had a stroke (Yes / No). |
| PhysicalHealth     | Continuous numerical    | Number of days (0–30) with poor physical health. |
| MentalHealth       | Continuous numerical    | Number of days (0–30) with poor mental health. |
| DiffWalking        | Binary                  | Indicates difficulty walking (Yes / No). |
| Sex                | Nominal categorical     | Categories with no natural order (Male / Female). |
| AgeCategory        | Ordinal categorical     | Age groups have a natural order from younger to older. |
| Race               | Nominal categorical     | Categories with no inherent order. |
| Diabetic           | Nominal categorical     | Multiple categories without strict order (Yes / No / Borderline). |
| PhysicalActivity   | Binary                  | Indicates physical activity (Yes / No). |
| GenHealth          | Ordinal categorical     | Ordered health levels (Poor → Excellent). |
| SleepTime          | Continuous numerical    | Numeric values representing hours of sleep. |
| Asthma             | Binary                  | Indicates whether the person has asthma (Yes / No). |
| KidneyDisease      | Binary                  | Indicates chronic kidney disease (Yes / No). |
| SkinCancer         | Binary                  | Indicates skin cancer diagnosis (Yes / No). |

### 11- Encoding Strategy

```python
df.head()
```

```python
df['HeartDisease'] = df['HeartDisease'].map({'Yes': 1, 'No': 0})
df['Smoking'] = df['Smoking'].map({'Yes': 1, 'No': 0})
df['AlcoholDrinking'] = df['AlcoholDrinking'].map({'Yes': 1, 'No': 0})
df['Stroke'] = df['Stroke'].map({'Yes': 1, 'No': 0})
df['DiffWalking'] = df['DiffWalking'].map({'Yes': 1, 'No': 0})
df['Sex'] = df['Sex'].map({'Male': 1, 'Female': 0})
df['Diabetic'] = df['Diabetic'].map({'No': 0, 'Yes (during pregnancy)': 1, 'No, borderline diabetes': 2, 'Yes': 3})
df['PhysicalActivity'] = df['PhysicalActivity'].map({'Yes': 1, 'No': 0})
df['GenHealth'] = df['GenHealth'].map({'Excellent': 4, 'Very good': 3, 'Good': 2, 'Fair': 1, 'Poor': 0})
df['Asthma'] = df['Asthma'].map({'Yes': 1, 'No': 0})
df['KidneyDisease'] = df['KidneyDisease'].map({'Yes': 1, 'No': 0})
df['SkinCancer'] = df['SkinCancer'].map({'Yes': 1, 'No': 0})
age_order = {
    '18-24': 0,
    '25-29': 1,
    '30-34': 2,
    '35-39': 3,
    '40-44': 4,
    '45-49': 5,
    '50-54': 6,
    '55-59': 7,
    '60-64': 8,
    '65-69': 9,
    '70-74': 10,
    '75-79': 11,
    '80 or older': 12
   
}
df["AgeCategory"] = df["AgeCategory"].map(age_order)
from sklearn.preprocessing import OneHotEncoder
encode = OneHotEncoder(sparse_output=False)
encode_col = encode.fit_transform(df[["Race"]])
encode_df = pd.DataFrame(encode_col, columns=encode.get_feature_names_out(["Race"]))
df_heatmap = df.reset_index(drop=True)
df_encoded = pd.concat([df.drop("Race", axis=1), encode_df], axis=1)
df_encoded.head(10)

```

```python
numeric_cols = ['BMI','PhysicalHealth'	,'MentalHealth','SleepTime']
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

df_no_outliers = df[~((df[numeric_cols] < (Q1 - 1.5*IQR)) | (df[numeric_cols] > (Q3 + 1.5*IQR))).any(axis=1)]
```

```python
df_no_outliers.info()
```

```python
corr = df_no_outliers[df_no_outliers.columns.difference(["Race"])].corr()
corr.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

```python
target_corr = corr['HeartDisease'].drop('HeartDisease').sort_values(ascending=False)

decision_df = pd.DataFrame({
    'Feature': target_corr.index,
    'Correlation': target_corr.values,
    'Decision': ['Strong' if abs(x) > 0.2 else 'Medium' if abs(x) > 0.1 else 'Weak' for x in target_corr]
})
```

```python
def style_decision_table(df):
    return df.style.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#1a5b7a'), ('color', 'white'), 
                                    ('font-family', 'Arial'), ('text-align', 'center'), ('padding', '10px')]},
        {'selector': 'td', 'props': [('background-color', '#d9effc'), ('color', 'black'), 
                                    ('border', '1px solid #7ebed9'), ('padding', '8px')]}
    ]).hide(axis='index').background_gradient(subset=['Correlation'], cmap='Blues')

style_decision_table(decision_df)
```
